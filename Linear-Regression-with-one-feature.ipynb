### Import library

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Training Data

```python
x_train=np.array([1,2])
y_train=np.array([300,500])

print(f"x_train= {x_train}")
print(f"y_train= {y_train}")
```

```python
print(f"x_train.sahp ={x_train.shape}")
# number of training example
m=x_train.shape[0]
print(f"number of training example is {m}")
```

```python
# another way to get number of training example is len(x_train)
m=len(x_train)
print(m)
```

```python
# training set examples
for i in range(m):
    x_train_i=x_train[i]
    y_train_i=y_train[i]

    print(f"(x^({i}),y^({i})=({x_train_i},{y_train_i})")
```

### Plot data without prediction

```python
# plot the data point
plt.scatter(x_train,y_train,marker="x",c='r')
# set title 
plt.title("House priecs")
# set y_axis lable
plt.ylabel("price in (1000s dollars)")
# set x_axis label
plt.xlabel("size in 1000 sqrft")
plt.show()

```

### Model Function

```python
w=200
b=100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(w,b,x):
    m=x.shape[0]
    f=np.zeros(m)
    
    for i in range(m):
        f[i]=w*x[i]+b
        
    return f
    
    
    
    
```

### Plot data befor and after the predition

```python
f_wb=compute_model_output(w,b,x_train)

plt.plot(x_train,f_wb,marker="x",c='b',label='prediction value')
plt.plot(x_train,y_train,marker="x",c="r",label="Actual value")
# set title 
plt.title("House priecs")
# set y_axis lable
plt.ylabel("price in (1000s dollars)")
# set x_axis label
plt.xlabel("size in 1000 sqrft")
plt.legend()
plt.show()
```
